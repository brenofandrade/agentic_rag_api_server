```python
import requests
import json
from IPython.display import display, Markdown
import uuid
```

```python
url = "http://localhost:8000/health"

r = requests.get(url)

if r.status_code == 200:
    print("SUCESSO")
else:
    print(r.status_code)
    print(r.content)
```

```python
def ask(question:str):

    url = "http://localhost:8000/chat"
    
    session_id = str(uuid.uuid4())
    # question = "When was build the Notre Dame University?"
    
    payload = {
        "question":question,
        "session_id":session_id
    }
    
    r = requests.post(url, json=payload, timeout=300)
    resposta = json.loads(r.content).get('answer')

    display(Markdown(resposta))

    return resposta

```

```python
ask("Olá, tudo bem?")
```

```python
greetings_test = [
    "Oi",
    "Olá",
    "Como vai?",
    "Bom dia!",
    "Boa tarde",
    "Boa noite"
]

for msg in greetings_test:
    resp = ask(msg)
    
```

```python
farewell_test = [
    "Adeus",
    "Tchau",
    "obrigado",
    "obrigado pela ajuda",
    "Valeu",
    "Até mais"
]

for msg in farewell_test:
    _ = ask(msg)
```
